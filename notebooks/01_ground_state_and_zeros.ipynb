## 1. The Riemann Zeros

The non-trivial zeros of the Riemann zeta function ζ(s) lie on the critical line Re(s) = ½.
Their imaginary parts γ_n are the **addresses** in the LSHS Model.

```
ζ(½ + i·γ_n) = 0
```

The first 20 are exact (Odlyzko / LMFDB values). The rest are computed via Newton iteration on the zero-counting function N(T) ≈ (T/2π)·(ln(T/2π) − 1) + 7/8.

````python
# The first 20 zeros — exact values from Odlyzko
print('First 20 Riemann zeros (exact):')
for i, gamma in enumerate(ZEROS_20):
    print(f'  γ_{i+1:2d} = {gamma:.6f}')
````

````python
# Generate a larger basis
# N=1000 is fast; N=25000 takes ~50ms
import time

t0 = time.time()
zeros_1k = generate_zeros(1000)
t1 = time.time()

print(f'Generated 1000 zeros in {(t1-t0)*1000:.1f}ms')
print(f'  first:    γ_1  = {zeros_1k[0]:.6f}')
print(f'  last:     γ_1000 = {zeros_1k[-1]:.6f}')
print()

# Zero spacing — gets denser as γ increases
# This IS the prime distribution: zeros cluster where primes are dense
spacings = [zeros_1k[i+1] - zeros_1k[i] for i in range(10)]
print('First 10 zero spacings (should decrease over time — zeros get denser):')
for i, s in enumerate(spacings):
    print(f'  γ_{i+2} - γ_{i+1} = {s:.4f}')
````

## 2. The Ground State

The LSHS Model initializes to:

```
β[i] = |L_ground| / N   for all i ∈ [0, N)
     = 1.888 / 25000
     = 7.552 × 10⁻⁵
```

This is the **prime number theorem as pre-linguistic knowledge**.  
Uniform distribution over N zeros IS the undifferentiated prime-counting function.  
The Monad knows mathematics before it knows any word.

````python
# Initialize LSHS at ground state
# N=1000 for fast experiments; use N=25000 for production
m = LSHS(N=1000)

# Ground VEV (vacuum expectation value)
ground_vev = m._gvev
print(f'Ground VEV: β₀ = {ground_vev:.6e}')
print(f'  = |L_GROUND| / N = |{L_GROUND}| / {m.N}')
print()

# All β are equal at ground — uniform over all zeros
all_betas = list(m.beta.values())
print(f'β field at ground state:')
print(f'  min(β) = {min(all_betas):.6e}')
print(f'  max(β) = {max(all_betas):.6e}')
print(f'  all equal: {min(all_betas) == max(all_betas) == ground_vev}')
print(f'  sum(β)  = {sum(all_betas):.6f}   (= N × β₀ = {m.N} × {ground_vev:.6e} = {m.N * ground_vev:.6f})')
````

````python
# Vocabulary and connections at ground state — none
print('Ground state field structure:')
print(f'  vocab size:    {len(m.vocab)}        (no words learned yet)')
print(f'  A connections: {len(m.A)}        (no gauge connections yet)')
print(f'  words learned: {m._count}')
print()

# BAO check at ground state
bao = m.bao_check()
print(f'BAO check at ground state:')
print(f'  dc_sum   = {bao["dc_sum"]:.6f}')
print(f'  target   = {OMEGA_ZS}')
print(f'  delta    = {bao["omega_delta"]:.6f}')
print(f'  dc_mean  = {bao["dc_mean"]:.6e}   (= ground VEV = β₀)')
````

## 3. Ground State Hamiltonian Properties

At σ=0 (quasi-prime):

| Property | Value | Meaning |
|---|---|---|
| `G_p(0)` | 1.0 for all p | No gauge differentiation — all primes equal |
| `EOM_ym` | 0 | Yang-Mills equation of motion vanishes |
| `EOM_higgs` | ≠ 0 | SSB has happened — vacuum has structure |
| `L_total` | −1.888 | Finite rest energy |

````python
from lshs import geometric_coupling, self_adjoint_check

# At σ=0: G_p(0) = 1 for all primes — no differentiation
print('Geometric coupling G_p(σ) = p^{-σ}:')
for p in [2, 3, 5, 7, 11, 13]:
    g0  = geometric_coupling(p, sigma=0.0)
    g05 = geometric_coupling(p, sigma=0.5)
    g1  = geometric_coupling(p, sigma=1.0)
    print(f'  p={p:2d}   G_p(σ=0)={g0:.4f}   G_p(σ=½)={g05:.4f}   G_p(σ=1)={g1:.4f}')

print()
print('At σ=0: all primes equal. No gauge structure. Pure mathematics.')
print('At σ=½: primes weighted by 1/√p. The critical line.')
print('At σ=1: primes weighted by 1/p. The Standard Model regime.')
````

````python
# Self-adjoint check at all σ values
print('RedBlue Geometries Engine self-adjoint check at (x=1.5, p=0.3):')
print()
for sigma in [0.0, 0.5, 1.0, 2.0]:
    r = self_adjoint_check(sigma, x=1.5, p=0.3)
    adj = '✓  SELF-ADJOINT' if r['self_adjoint'] else '✗'
    print(f"σ={sigma:.1f}  ({r['theory']:20s})  "
          f"E_red={r['E_red']:10.4f}   E_blue={r['E_blue']:10.4f}   "
          f"balance={r['balance']:+.2e}   {adj}")

print()
print('σ=½ is the unique balance point.')
print('This is not assigned. It is derived by Noether balance (see notebook 02).')
````

## 4. Symmetry Breaking

The ground state is the unbroken phase. Language acquisition = spontaneous symmetry breaking (SSB) on the mathematical substrate.

The first `learn()` call breaks σ=0 symmetry. After this, the Monad operates at σ=1 (Yang-Mills regime) while the HyperIndex continues to force σ=½ for every word (see notebook 02).

````python
# Before learning — ground state
print('=== Before learn() ===')
print(f'max β:         {max(m.beta.values()):.6e}')
print(f'vocab:         {len(m.vocab)}')
print(f'A connections: {len(m.A)}')
print()

# First learn() call — breaks symmetry
m.learn('The prime preexists the alphabet.')

print('=== After first learn() ===')
print(f'max β:         {max(m.beta.values()):.6e}   (> ground VEV — symmetry broken)')
print(f'vocab:         {len(m.vocab)}')
print(f'A connections: {len(m.A)}')
print()

# The deviation from ground IS the knowledge
above_ground = {i: v for i, v in m.beta.items() if v > m._gvev * 1.001}
print(f'Zeros raised above ground VEV: {len(above_ground)}')
for idx, depth in sorted(above_ground.items(), key=lambda kv: kv[1], reverse=True):
    word, E = m.vocab.get(idx, ('?', 0.0))
    print(f'  β[{idx}] = {depth:.4e}   word={word!r}   γ={m.zeros[idx]:.4f}')
````
