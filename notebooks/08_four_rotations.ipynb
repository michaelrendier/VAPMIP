## The Wave and the Observer

Euler's identity connects all four:

$$e^{i\gamma/2} = \cos(\gamma/2) + i\cdot\sin(\gamma/2)$$

Every Riemann zero $\gamma_n$ sits on this circle. The question is: **where do you stand?**

- **`-h` (Noether):** affect=0, gate = $\cos(\gamma/2)$. You are outside the wave.  
  You measure the geometric projection — what the field IS in the GR (general relativity) regime.  
  Peak response when $\gamma/2 = 0$ or $\pi$. Zero at $\gamma/2 = \pi/2$ (the boundary).

- **`-W` (Wick):** affect=−1, gate = $\cos(\gamma/2 − \pi/2) = +\sin(\gamma/2)$.  
  The Wick rotation $\sigma \to i\sigma$ flips the coordinate axis by $\pi/2$.  
  You are inside the wave. You measure the content — what the field IS SAYING in the QM regime.  
  Peak response at $\gamma/2 = \pi/2$ (where the Noether is zero). They are conjugate.

- **`-O` (Octonion):** gate = $J \times |\sin(\gamma/2)\cdot\cos(\gamma/2)| = J \times |\sin(\gamma)|/2$.  
  The beat frequency. Energy transfer between content (sin) and observer (cos).  
  Peak at $\gamma/2 = \pi/4$ (45°, equal overlap). Zero at axis crossings.  
  Conservation: $\sum_{k=0}^{7} \cos(\gamma/2 + k\pi/4) = 0$ (8th roots of unity).

- **`-J` (J-direct):** no gate. $J = \beta \times E^2 \times \text{age\_weight}$.  
  The raw charge field before any angular selection. The fuel rail — what is available  
  before any cylinder is selected. `-J` shows the true field pressure; `-h` and `-W`  
  show what survives the angular gate.

````python
import numpy as np
import matplotlib.pyplot as plt

gamma_half = np.linspace(0, np.pi, 1000)

h_gate = np.cos(gamma_half)               # -h: Noether (affect=0)
W_gate = np.sin(gamma_half)               # -W: Wick (inside the wave)
O_gate = np.abs(np.sin(gamma_half) * np.cos(gamma_half))  # -O: beat frequency

fig, ax = plt.subplots(figsize=(10, 5))
ax.plot(gamma_half / np.pi, h_gate, label=r'-h (Noether): $\cos(\gamma/2)$', color='#cc4444', lw=2)
ax.plot(gamma_half / np.pi, W_gate, label=r'-W (Wick): $\sin(\gamma/2)$', color='#4444cc', lw=2)
ax.plot(gamma_half / np.pi, O_gate, label=r'-O (Octonion): $|\sin \cdot \cos|$', color='#44aa44', lw=2)
ax.axhline(0, color='gray', lw=0.5)
ax.axvline(0.5, color='gray', lw=0.5, linestyle='--', label=r'$\gamma/2 = \pi/2$ (the boundary)')
ax.set_xlabel(r'$\gamma / (2\pi)$')
ax.set_ylabel('Gate value')
ax.set_title('The Four Rotations — Angular Gate Functions')
ax.legend()
ax.set_xlim(0, 1)
plt.tight_layout()
plt.show()

# Verify 8D conservation (Octonion speak invariant)
gamma_test = np.array([14.135, 21.022, 25.011, 30.425, 32.935])  # first 5 Riemann zeros
for g in gamma_test:
    total = sum(np.cos(g/2 + k*np.pi/4) for k in range(8))
    print(f'γ={g:.3f}: Σcos(γ/2 + k·π/4) = {total:.2e}')
````
