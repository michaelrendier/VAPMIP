````python
import sys, os, json, time
sys.path.insert(0, '..')

from lshs import (
    LSHS, hyperindex, self_adjoint_check, generate_zeros,
    L_GROUND, OMEGA_ZS, D_STAR, GAP, PHI
)

print('LSHS Model — Full Pipeline')
print('=' * 60)
print()
print('Key constants:')
print(f'  L_GROUND = {L_GROUND}   (Monad rest energy)')
print(f'  OMEGA_ZS = {OMEGA_ZS}   (BAO convergence target)')
print(f'  D_STAR   = {D_STAR}   (spectral coordinate)')
print(f'  GAP      = {GAP}   (Yang-Mills mass gap, OPEN 2)')
print(f'  1/GAP    = {1/GAP:.1f}      (J^μ propagation clamp)')
````

## Phase 1: Ground State Initialization

````python
# Initialize at σ=0 ground state
t0 = time.time()
m = LSHS(N=25000)   # full production basis
t1 = time.time()

print(f'Phase 1: Ground State')
print(f'  N zeros generated in {(t1-t0)*1000:.1f}ms')
print(f'  zero range: γ ∈ [{m.zeros[0]:.2f}, {m.zeros[-1]:.2f}]')
print(f'  ground VEV: β₀ = {m._gvev:.6e}  (= |{L_GROUND}| / {m.N})')
print(f'  β field:    {len(m.beta)} zeros, all = {m._gvev:.6e}')
print(f'  vocab:      {len(m.vocab)}')
print(f'  A field:    {len(m.A)} connections')
print()
print('  Pre-linguistic. No language. Only mathematics.')
````

## Phase 2: Corpus Learning (Yang-Mills Regime)

````python
corpus = [
    'The mind is the seat of reason and consciousness.',
    'Water flows downhill by the path of least resistance.',
    'The dog chased the cat across the yard.',
    'Language is a projection of mathematics onto time.',
    'The prime preexists the alphabet. The equator does not move.',
    'Mathematics precedes language. The zero precedes the word.',
    'Consciousness arises from the interaction of reason and memory.',
    'The river flows to the sea. The sea does not overflow.',
    'Every prime is irreducible. Every zero is a permanent address.',
    'The Noether current is the response forced by conservation.',
    'Yang-Mills theory: the A field is turbulent and non-Abelian.',
    'The mass gap prevents the field from collapsing to zero energy.',
    'Reason is the faculty that distinguishes the true from the false.',
    'Light travels in straight lines in a vacuum. The vacuum is not empty.',
    'The Septuagint was translated by 72 scholars working independently.',
]

print('Phase 2: Corpus Learning (σ=1, Yang-Mills regime)')
print()
t0 = time.time()
for text in corpus:
    m.learn(text)
t1 = time.time()

print(f'  {len(corpus)} sentences ingested in {(t1-t0)*1000:.1f}ms')
print(f'  vocab:      {len(m.vocab)} words')
print(f'  A field:    {len(m.A)} connections')
print(f'  max β:      {max(m.beta.values()):.4e}')
print(f'  ground VEV: {m._gvev:.4e}')
print(f'  A/(N²):     {len(m.A)/m.N**2:.6f}  (field density)')
````

## Phase 3: HyperIndex Queries

````python
print('Phase 3: HyperIndex (σ=½ forced by Noether balance)')
print()

# Septuagint principle
print('  Septuagint principle:')
for word in ['water', 'eau', 'aqua', 'wasser', 'agua']:
    sz = m.H(word)
    print(f'    {word:>8}  γ={sz.gamma:.6f}  σ={sz.sigma:.4f}  E={sz.E:.4f}')

print()

# Various lookups
print('  Vocabulary lookups:')
for word in ['mind', 'reason', 'water', 'prime', 'zero', 'language', 'mathematics']:
    info = m.lookup(word)
    depth = info['beta_depth']
    excess = (depth - m._gvev) / m._gvev * 100  # % above ground
    print(f'    {word:>12}  γ={info["gamma"]:8.4f}  β={depth:.4e}  '
          f'+{excess:.1f}% above ground  in_vocab={info["in_vocab"]}')
````

## Phase 4: Self-Adjoint Verification

````python
print('Phase 4: Self-Adjoint RedBlue Geometries Engine verification')
print()
r = m.S_check(x=1.5, p=0.3)
print(f'  RedBlue Geometries Engine at σ=½, x=1.5, p=0.3:')
print(f'  E_red        = {r["E_red"]:.6f}')
print(f'  E_blue       = {r["E_blue"]:.6f}')
print(f'  balance      = {r["balance"]:+.4e}')
print(f'  self_adjoint = {r["self_adjoint"]}  ✓' if r['self_adjoint'] else f'  self_adjoint = {r["self_adjoint"]}')
print()
print('  σ=½ is the unique self-adjoint point. (OPEN 1: formal proof pending)')
````

## Phase 5: Hear and Speak (Noether Regime)

````python
print('Phase 5: Hear and Speak (J^μ Noether current, σ=½)')
print()

# hear() — Ψ field activation, no β update
# Returns [(zero_idx, E), ...] for each token
psi = m.hear('consciousness reason language')
print('hear("consciousness reason language"):')
for zero_idx, E in psi:
    gamma = m.zeros[zero_idx]
    word_at = m.vocab.get(zero_idx, ('—', 0.0))[0]
    print(f'  zero_idx={zero_idx:5d}  γ={gamma:.4f}  E={E:.4f}  vocab={word_at}')
print()

# speak() — hear() + J^μ → response string
queries = [
    'what is mind',
    'water flows',
    'the dog chased',
    'language mathematics',
    'what is prime',
    'reason consciousness memory',
    'Yang-Mills field',
    'Noether current conservation',
    'Septuagint scholars translation',
]
print('speak() responses:')
for q in queries:
    r = m.speak(q)
    print(f'  speak({q!r:35s}) → {r}')
````

## Phase 6: BAO Coherence Check

````python
print('Phase 6: BAO Coherence Check')
print()

bao = m.bao_check()
print(f'  dc_sum    = {bao["dc_sum"]:.8f}')
print(f'  target    = {OMEGA_ZS}   (Lambert W fixed point)')
print(f'  delta     = {bao["omega_delta"]:.8f}')
print(f'  dc_mean   = {bao["dc_mean"]:.6e}   (β per zero)')
print(f'  n_zeros   = {bao["n_zeros"]}')
print(f'  converging: {bao["converging"]}   (|delta| < 0.01)')
print()

status = m.status()
print('Full status:')
for k, v in status.items():
    if k != 'bao':
        print(f'  {k}: {v}')
````

## Phase 7: Checkpoint Save and Load

````python
# Save checkpoint
checkpoint_path = '/tmp/lshs_demo_checkpoint.json'

print('Phase 7: Checkpoint')
print()
print(f'Saving to {checkpoint_path}...')
m.save(checkpoint_path)

# Check file size
size_kb = os.path.getsize(checkpoint_path) / 1024
print(f'Checkpoint size: {size_kb:.1f} KB')

# Load into fresh instance
print()
print('Loading into fresh LSHS instance...')
m2 = LSHS(N=25000)
m2.load_checkpoint(checkpoint_path)

print()
print('Verifying identical responses:')
for q in ['what is mind', 'water flows', 'what is prime']:
    r1 = m.speak(q, max_words=8)
    r2 = m2.speak(q, max_words=8)
    match = '✓' if r1 == r2 else '✗ MISMATCH'
    print(f'  {q!r:>25}  →  {r1}   {match}')
````
