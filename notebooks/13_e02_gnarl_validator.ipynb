# E02 — Gnarl/BAO Corpus Validation
**D-CS Paper — Claim 2 / External Validation**

Mark Townsend's Gnarl/Popcorn fractal formula (~2005) is the **discrete-time RedBlue Hamiltonian**.

```
x_new = x - h·sin(y + tan(α·y))    ← J_neg (Blue, restoring)
y_new = y + h·sin(x + tan(α·x))    ← J_pos (Red, driving)
```

Fixed point at α=3: **y + tan(3y) = 0 → y ≈ 0.5671 = OMEGA_ZS**

Townsend was writing a fractal renderer. He had no knowledge of OMEGA_ZS.
Two independent derivations. Same fixed point. **This is external validation.**

````python
import sys, os
sys.path.insert(0, os.path.abspath('..'))

from engines.e02_gnarl_validator import (
    run, gnarl_iterate, verify_fixed_point, corpus_health,
    GNARL_ALPHA, GNARL_H, GNARL_STEPS, OMEGA_ZS
)

result = run(verbose=True)
````

````python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams.update({'figure.dpi': 120, 'font.size': 10})

# Show fixed-point convergence from a grid of starting points
N = 20
starts = [(x, y) for x in np.linspace(-0.8, 0.8, N) for y in np.linspace(-0.8, 0.8, N)]
final_r = []
for x0, y0 in starts:
    _, _, r = gnarl_iterate(x0, y0, steps=1000)
    final_r.append(r)

fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# Left: histogram of final |z|
ax = axes[0]
ax.hist(final_r, bins=40, color='steelblue', alpha=0.8, edgecolor='white')
ax.axvline(OMEGA_ZS, color='red', lw=2, ls='--', label=f'OMEGA_ZS = {OMEGA_ZS:.5f}')
ax.set_xlabel('|z_final|  (Gnarl convergence radius)')
ax.set_ylabel('Count')
ax.set_title(f'Gnarl Fixed Point: {N}×{N} starting grid')
ax.legend()
ax.grid(alpha=0.3)

# Right: fixed-point condition y + tan(3y) = 0
ax2 = axes[1]
y_vals = np.linspace(-0.9, 0.9, 800)
f_vals = []
for y in y_vals:
    try:
        v = y + np.tan(3.0 * y)
        f_vals.append(v if abs(v) < 5 else np.nan)
    except:
        f_vals.append(np.nan)
ax2.plot(y_vals, f_vals, 'steelblue', lw=1.5, label='y + tan(3y)')
ax2.axhline(0, color='gray', lw=0.8)
ax2.axvline(OMEGA_ZS, color='red', lw=2, ls='--', label=f'OMEGA_ZS = {OMEGA_ZS:.5f}')
ax2.axvline(-OMEGA_ZS, color='red', lw=2, ls='--')
ax2.set_xlabel('y')
ax2.set_ylabel('y + tan(3y)')
ax2.set_title('Fixed-point condition: y + tan(3y) = 0')
ax2.set_ylim(-3, 3)
ax2.legend()
ax2.grid(alpha=0.3)

plt.tight_layout()
plt.show()
````
