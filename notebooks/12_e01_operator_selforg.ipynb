# E01 — Operator Self-Organisation
**D-CS Paper — Claim 1**

16 sedenion operator names, chosen for semantic reasons, prime-hash to three geometric zones:
- **GROUND** (E ≈ d* = 0.246): minimum-energy operators
- **CRITICAL** (E ≈ σ½ = 0.5): critical-line operators  
- **BOUNDARY** (E ≈ D* = 1.0): zero-divisor boundary operators

**Zero free parameters. No training. No fitting.**

The result: `d* / σ_mean / D* = 1`

```python
import sys, os
sys.path.insert(0, os.path.abspath('..'))

from engines.e01_operator_selforg import run, OPERATORS, D_STAR_VAL, SIGMA_HALF, D_BOUNDARY

result = run(verbose=True)
```

```python
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
plt.rcParams.update({'figure.dpi': 120, 'font.size': 10})

results = result['results']
sorted_r = sorted(results, key=lambda x: -x['E'])

names = [r['op'] for r in sorted_r]
energies = [r['E'] for r in sorted_r]
zone_colors = {'BOUNDARY': '#d62728', 'CRITICAL': '#ff7f0e', 'GROUND': '#1f77b4'}
colors = [zone_colors[r['zone']] for r in sorted_r]

fig, ax = plt.subplots(figsize=(9, 6))
bars = ax.barh(range(len(names)), energies, color=colors, alpha=0.85, edgecolor='white')

ax.axvline(D_STAR_VAL, color='#1f77b4', ls='--', lw=1.5, label=f'd* = {D_STAR_VAL}')
ax.axvline(SIGMA_HALF, color='#ff7f0e', ls='--', lw=1.5, label=f'σ½ = {SIGMA_HALF}')
ax.axvline(D_BOUNDARY, color='#d62728', ls='--', lw=1.5, label=f'D* = {D_BOUNDARY}')

ax.set_yticks(range(len(names)))
ax.set_yticklabels(names)
ax.set_xlabel('Prime-hash Energy E')
ax.set_title('16 Operator Names → Sedenion Geometric Zones\n(zero free parameters)')

patches = [mpatches.Patch(color=c, label=z) for z, c in zone_colors.items()]
ax.legend(handles=patches + ax.get_lines(), loc='lower right', fontsize=9)
ax.grid(axis='x', alpha=0.3)
plt.tight_layout()
plt.show()

print(f"\nGROUND  mean = {result['g_mean']:.4f}  target d* = {D_STAR_VAL}  ratio = {result['g_convergence']:.4f}")
print(f"CRITICAL mean = {result['c_mean']:.4f}  target σ½ = {SIGMA_HALF}  ratio = {result['c_convergence']:.4f}")
print(f"BOUNDARY mean = {result['b_mean']:.4f}  target D* = {D_BOUNDARY}  ratio = {result['b_convergence']:.4f}")
print(f"\nd* / σ_mean / D* = {D_STAR_VAL / result['c_mean'] / D_BOUNDARY:.6f}  (target = 1.0)")
```
