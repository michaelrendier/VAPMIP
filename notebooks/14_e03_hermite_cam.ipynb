# E03 — Hermite H₁₆ CAM Timing Wheel
**D-CS Paper — Claim 3**

The 16th-degree Hermite polynomial has exactly **16 real zeros**.
Their spacing is **GUE-distributed** — same statistics as Riemann zero spacing.

Assignment: sedenion dimension e_k is resonant at the k-th zero of H₁₆(z).

This is the **CAM timing wheel calibration** — the sedenion equivalent of a diesel
engine's camshaft profile. Uniform E-values = untrained engine. Hermite-spaced = calibrated.

**External validation:** Dave Makin's Transpoly at degree 16 (mmf.ufm, Ultra Fractal)
independently produces this 16-petal structure — one petal per Hermite zero.
Makin was rendering fractals.

```python
import sys, os
sys.path.insert(0, os.path.abspath('..'))

from engines.e03_hermite_cam import run, hermite_zeros_numpy, gue_spacing_statistics, OPERATORS, OMEGA_ZS

result = run(verbose=True)
```

```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams.update({'figure.dpi': 120, 'font.size': 10})

zeros   = result['zeros']
targets = result['e_targets']
stats   = result['stats']

fig, axes = plt.subplots(1, 3, figsize=(14, 5))

# Left: H₁₆ zero positions
ax = axes[0]
ax.scatter(zeros, [0]*16, c='steelblue', s=80, zorder=3)
ax.axhline(0, color='gray', lw=0.8)
for k, (z, op) in enumerate(zip(zeros, OPERATORS)):
    ax.annotate(f'e{k}', (z, 0), textcoords='offset points',
                xytext=(0, 10 if k%2==0 else -18), fontsize=7, ha='center')
ax.set_xlabel('H₁₆ zero position')
ax.set_title('H₁₆: 16 Real Zeros\n(sedenion CAM calibration points)')
ax.set_yticks([])
ax.grid(axis='x', alpha=0.3)

# Middle: GUE spacing distribution
ax2 = axes[1]
gaps = stats['gaps']
mean_gap = stats['mean_gap']
s = [g / mean_gap for g in gaps]
ax2.hist(s, bins=8, color='steelblue', alpha=0.8, edgecolor='white', density=True)
# Wigner surmise (GUE)
ss = np.linspace(0, 3, 200)
wigner = (32/np.pi**2) * ss**2 * np.exp(-4*ss**2/np.pi)
ax2.plot(ss, wigner, 'r--', lw=2, label='GUE Wigner surmise')
ax2.set_xlabel('Normalised spacing s')
ax2.set_ylabel('Density')
ax2.set_title(f'Spacing statistics\nWigner-Dyson ratio = {stats["wd_ratio"]:.3f}')
ax2.legend(fontsize=8)
ax2.grid(alpha=0.3)

# Right: E-value targets vs uniform
ax3 = axes[2]
dims = range(16)
ax3.plot(dims, targets, 'o-', color='steelblue', label='Hermite-calibrated', lw=2)
ax3.axhline(OMEGA_ZS/2, color='gray', ls='--', lw=1.2, label=f'Uniform (OMEGA_ZS/2 = {OMEGA_ZS/2:.3f})')
ax3.axhline(OMEGA_ZS, color='red', ls=':', lw=1.2, label=f'OMEGA_ZS = {OMEGA_ZS:.4f}')
ax3.set_xlabel('Sedenion dimension')
ax3.set_ylabel('E-value target')
ax3.set_title('CAM Calibration Targets\nHermite vs uniform')
ax3.set_xticks(range(16))
ax3.set_xticklabels([f'e{k}' for k in range(16)], rotation=45, fontsize=7)
ax3.legend(fontsize=8)
ax3.grid(alpha=0.3)

plt.tight_layout()
plt.show()
```
